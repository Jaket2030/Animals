### Part2 : Running Decision Tree

```python
new_df = pd.get_dummies(df, columns=['AnimalType','SexuponOutcome'])
new_df.head
new_df_train, new_df_test = split(new_df, test_size=0.3, random_state=4444)
new_df.head()
```

```python
#Fit the model
outcome_dt = DecisionTreeClassifier(max_depth=5,
                                      min_samples_split=10,
                                      min_samples_leaf=10)

X = new_df_train.drop(['DateTime','OutcomeSubtype','AgeuponOutcome','COUNTER','Age','A','B','COUNTER_Outcome','Color','Breed','OutcomeType','Name'], axis=1)
y = new_df_train.OutcomeType
#Checking if I have Nulls in the Data
pd.isnull(new_df_train).sum() > 0
new_df_train.fillna(0, inplace=True)
pd.isnull(new_df_train).sum() > 0
outcome_dt.fit(X, y)
```

```python
#Converting the outcome classes
print(outcome_dt.classes_)
```

```python
#Exporting the tree image using graphviz
export_graphviz(decision_tree=outcome_dt,
                out_file='outcome.dot',
                feature_names=X.columns,
                class_names=[str(x) for x in outcome_dt.classes_], 
                leaves_parallel=True,
                filled=True,
                rotate=False,
                rounded=True)
```

```python
#Printing the confusion matrix
cm = confusion_matrix(y_true=new_df_train.OutcomeType,
                      y_pred=outcome_dt.predict(X),
                      labels=outcome_dt.classes_)
pd.DataFrame(cm,
             index=outcome_dt.classes_,
             columns=outcome_dt.classes_)
```

```python
#Classification Report
print(classification_report(y_true=new_df_train.OutcomeType,
                      y_pred=outcome_dt.predict(X)))
```
